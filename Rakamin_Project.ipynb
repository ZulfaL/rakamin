```python

import pandas as pd
data_test = pd.read_csv('application_test.csv')
data_test.head(10)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import math
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
dataset = data_train[["SK_ID_CURR","TARGET","CNT_FAM_MEMBERS"]]
dataset
```

```python
print("jumlah dataset: "+str(len(dataset.index)))
```

```python
dataset.describe()
```

```python
dataset.isna().sum() #cek missing value
```

##PENANGANAN MISSING VALUE

```python
# Menghitung median untuk setiap kolom
median_per_kolom = dataset.median()

# Mengisi nilai kosong dengan median dari setiap kolom
data = dataset.fillna(median_per_kolom)

data
```

```python
data.isna().sum() #cek missing value
```

```python
data.plot(x="CNT_FAM_MEMBERS",y="TARGET",style='o')
plt.title("Banyak Keluarga vs Kegagalan")
plt.xlabel("Banyak Keluarga")
plt.ylabel("Keberhasilan")
plt.show()
```

```python
# Visualisasi hubungan antara CNT_FAM_MEMBERS dan tingkat kegagalan pembayaran
plt.figure(figsize=(10, 6))
sns.barplot(x='CNT_FAM_MEMBERS', y='TARGET', data=data, ci=None)
plt.title('Hubungan antara CNT_FAM_MEMBERS dan Tingkat Kegagalan Pembayaran')
plt.xlabel('Jumlah Anggota Keluarga (CNT_FAM_MEMBERS)')
plt.ylabel('Rata-rata Tingkat Kegagalan Pembayaran')
plt.show()
```

```python

from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Memilih kolom yang akan dinormalisasi
columns_to_normalize = ['CNT_FAM_MEMBERS']

# Menggunakan MinMaxScaler
scaler = MinMaxScaler()
data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])

# Menyimpan dataset yang sudah dinormalisasi
#df.to_csv('dataset_normalized.csv', index=False)
```

```python
data
```

## Analisis Data

```python
# Pilih fitur (variabel independen) dan target (variabel dependen)
fitur = data[['CNT_FAM_MEMBERS']]  # Variabel independen
target = data['TARGET']  # Variabel dependen

# Bagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(fitur, target, test_size=0.2, random_state=42)

# Inisialisasi model regresi linier
model = LinearRegression()

# Latih model pada data latih
model.fit(X_train, y_train)

# Lakukan prediksi pada data uji
y_pred = model.predict(X_test)

# Evaluasi performa model menggunakan mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Tampilkan parameter regresi
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_)
```

```python
dataframe = pd.DataFrame({"Data Sebenarnya" : y_test, "Data Prediksi" : y_pred})
dataframe
```

```python
# Menampilkan visualisasi hasil prediksi
plt.scatter(X_test["CNT_FAM_MEMBERS"], y_test, color='black', label='Actual')
plt.scatter(X_test["CNT_FAM_MEMBERS"], y_pred, color='blue', label='Predicted')
plt.xlabel("CNT_FAM_MEMBERS")
plt.ylabel("TARGET")
plt.legend()
plt.show()
```
